```python
# Read in csv file
iowa_house = pd.read_csv('Files/IowaHouse.csv')
```

```python
# Create function to predict Iowa population in 2030 using the average census rate of change from 1990 to 2020
def predict_population_2030(census_1990, census_2000, census_2010, census_2020):
    rate_1990_2000 = (census_2000 - census_1990) / census_1990
    rate_2000_2010 = (census_2010 - census_2000) / census_2000
    rate_2010_2020 = (census_2020 - census_2010) / census_2010
    
    avg_rate_of_change = (rate_1990_2000 + rate_2000_2010 + rate_2010_2020) / 3
    
    predicted_population_2030 = census_2020 * (1 + avg_rate_of_change)
    
    return predicted_population_2030
```

```python
# Create new column with the predicted 2030 population per house district
iowa_house['pop_pred_2030'] = predict_population_2030(iowa_house.iloc[:,63], iowa_house.iloc[:,54], iowa_house.iloc[:,43], iowa_house.iloc[:,2])
```

```python
# Total predicted Iowa 2030 population
sum_pop_pred30 = iowa_house['pop_pred_2030'].sum()
print(sum_pop_pred30)
```
